```python
import pandas as pd
import folium
```

```python
df = pd.read_csv("../local-data/CuisineRetailLocations2.csv")
```

```python
df_uncategorized = df[df['CUISINE_TYPE'] == "Uncategorized"]
display(df_uncategorized['STORE_NAME'].unique())
```

```python
active_map = folium.Map(location=[32.7767, -96.7970], zoom_start=10)
cuisine_types = df['CUISINE_TYPE'].unique()

colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue',
          'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen',
          'gray', 'black', 'lightgray']

color_map = {cuisine: colors[i % len(colors)] for i, cuisine in enumerate(cuisine_types)}

feature_groups = {}
for cuisine in cuisine_types:
    feature_groups[cuisine] = folium.FeatureGroup(name=cuisine, show=False)

for idx, row in df.iterrows():
    if pd.notna(row['LATITUDE']) and pd.notna(row['LONGITUDE']): 
        cuisine = row['CUISINE_TYPE']
        popup_text = f"<b>{row['STORE_NAME']}</b><br>Type: {cuisine}"

        folium.Marker(
            location=[row['LATITUDE'], row['LONGITUDE']],
            popup=folium.Popup(popup_text, max_width=300),
            icon=folium.Icon(color=color_map[cuisine], icon='info-sign')
        ).add_to(feature_groups[cuisine])

for fg in feature_groups.values():
    fg.add_to(active_map)

folium.LayerControl().add_to(active_map)

active_map
```
