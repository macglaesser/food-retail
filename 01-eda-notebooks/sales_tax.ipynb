### Taxpayers who hold an active sales tax permit under Tax Code Chapter 151, Subchapter F

```python
import pandas as pd
import folium
```

```python
graph_df = pd.read_csv("../local-data/final_geocoded_output.csv")
graph_df.columns = graph_df.columns.str.upper().str.replace(' ', '_')
```

```python
graph_df.columns
```

```python
for col in graph_df.columns:
    print(f"{col}: {graph_df[col].isna().sum()} NaN values")
```

```python
print(graph_df.MATCH_STATUS.value_counts())
```

```python
print(graph_df.MATCH_TYPE.value_counts())
```

```python
graph_df = graph_df.dropna()
graph_df.columns
dallas_map = folium.Map(location=[32.7767, -96.7970], zoom_start=10)

for idx, row in graph_df.iterrows():
    folium.Marker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        popup=row['TAXPAYER_NAME']
    ).add_to(dallas_map)

dallas_map
```
