```python
snap['SNAP_COMPOSITE_KEY'] = (snap['STORE_STREET_ADDRESS'].astype(str) + " " + snap['CITY'].astype(str) + " " + snap['STATE'].astype(str))
tax['TAX_COMPOSITE_KEY'] = (tax['OUTLET_ADDRESS'] + " " +  tax['OUTLET_CITY'] + " " + tax['OUTLET_STATE'])
display(snap['SNAP_COMPOSITE_KEY'].head())
display(tax['TAX_COMPOSITE_KEY'].head())
```

```python
# 1. Unique key sets
snap_keys = set(snap['SNAP_COMPOSITE_KEY'])
tax_keys  = set(tax['TAX_COMPOSITE_KEY'])

# 2. Intersection and unions
common_keys     = snap_keys & tax_keys
only_snap_keys  = snap_keys - common_keys
only_tax_keys   = tax_keys  - common_keys

# 3. Row‐level match counts
tax_row_matches  = tax['TAX_COMPOSITE_KEY'].isin(snap_keys).sum()
snap_row_matches = snap['SNAP_COMPOSITE_KEY'].isin(tax_keys).sum()

# 4. Totals
total_tax = len(tax)
total_snap = len(snap)

# 5. Print them out
print(f"Unique keys in SNAP:           {len(snap_keys)}")
print(f"Unique keys in TAX:            {len(tax_keys)}")
print(f"Unique keys in BOTH:           {len(common_keys)}")
print(f"Unique only in SNAP:           {len(only_snap_keys)}")
print(f"Unique only in TAX:            {len(only_tax_keys)}\n")

print(f"Rows in TAX matching SNAP:     {tax_row_matches} of {total_tax}")
print(f"Rows in SNAP matching TAX:     {snap_row_matches} of {total_snap}")
print(f"TAX match rate (rows):         {tax_row_matches/total_tax:.1%}")
print(f"SNAP match rate (rows):        {snap_row_matches/total_snap:.1%}")
print(f"TAX match rate (unique keys):  {len(common_keys)/len(tax_keys):.1%}")
print(f"SNAP match rate (unique keys): {len(common_keys)/len(snap_keys):.1%}")
```

```python
# Shorter store type names
naics_to_store_type_short = {
    445110: 'Supermarket',
    445120: 'Convenience Store',
    445210: 'Meat Market',
    445220: 'Seafood Market',
    445230: 'Produce Market',
    445291: 'Bakery',
    445292: 'Confectionery',
    445299: 'Specialty Food',
    445310: 'Liquor Store',
    445000: 'Food Store',
    445100: 'Grocery Store'
}

tax['STORE_TYPE'] = tax['OUTLET_NAICS_CODE'].map(naics_to_store_type_short)
tax.head()
```

```python
tax_unmatched = tax.loc[
    ~tax['TAX_COMPOSITE_KEY'].isin(snap['SNAP_COMPOSITE_KEY'])
].copy()
```

```python
import uuid

def gen_id(): 
    # Generate a random 64-bit integer
    return uuid.uuid4().int >> 64

# 2a. New primary key
tax_unmatched['RECORD_ID'] = [gen_id() for _ in range(len(tax_unmatched))]

# 2b. Rename / map columns
tax_unmatched['STORE_NAME']           = tax_unmatched['OUTLET_NAME']
tax_unmatched['STORE_STREET_ADDRESS'] = tax_unmatched['OUTLET_ADDRESS']
tax_unmatched['Additonal_Address']    = ""  # no secondary address in TAX
tax_unmatched['CITY']                 = tax_unmatched['OUTLET_CITY']
tax_unmatched['STATE']                = tax_unmatched['OUTLET_STATE']
tax_unmatched['Zip_Code']             = tax_unmatched['OUTLET_ZIP_CODE']
tax_unmatched['Zip4']                 = ""  # blank if not available
tax_unmatched['County']               = tax_unmatched['OUTLET_COUNTY_CODE'].apply(lambda c: 'DALLAS' if int(c)==57 else 'Other')
# You already did this:
# tax_unmatched['STORE_TYPE']         = tax_unmatched['OUTLET_NAICS_CODE'].map(naics_to_store_type_short)
tax_unmatched['Latitude']             = tax_unmatched['LATITUDE']
tax_unmatched['Longitude']            = tax_unmatched['LONGITUDE']
tax_unmatched['Incentive_Program']    = ""   # no equivalent
tax_unmatched['Grantee_Name']         = ""   # no equivalent
tax_unmatched['ObjectId']             = None # or generate if needed
```

```python
# Ensure the new rows have the *exact same* columns, in the *same order*, as your snap DataFrame:
snap_cols = [
    'RECORD_ID',
    'STORE_NAME',
    'STORE_STREET_ADDRESS',
    'Additonal_Address',
    'CITY',
    'STATE',
    'Zip_Code',
    'Zip4',
    'County',
    'STORE_TYPE',
    'Latitude',
    'Longitude',
    'Incentive_Program',
    'Grantee_Name',
    'ObjectId'
]

new_snap_rows = tax_unmatched[snap_cols]
```

```python
snap_expanded = pd.concat([snap, new_snap_rows], ignore_index=True)
```
